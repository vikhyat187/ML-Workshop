```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = Image.open("thor.jfif")
img
```

```python

arr= np.array(img)
arr.shape
```

```python
work = arr.copy()
work.shape
```

```python
work[:,:,[1,2]]=0
plt.imshow(work)
```

```python
black_and_white = arr.copy()
black_and_white.shape
```

```python
#we are going to create a black and white filter
gray = arr.mean(axis=2)
plt.imshow(gray)
#axis =0 across row x axis 
#axis = 1 across column y axis 
#axis =2 across depth z axis
```

```python
#cmap = color mapping in prev one it's using multicolor
plt.imshow(gray,cmap="gray")
```

```python
cut = arr[-5:,-5:]
plt.imshow(cut)
```
